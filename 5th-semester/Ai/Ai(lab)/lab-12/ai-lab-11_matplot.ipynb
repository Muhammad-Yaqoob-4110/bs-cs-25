#### Make data using numpy


```python
# make data using numpy np.linspace(start, stop, num)
x = np.linspace(1, 10, 10)


y = x**2


z = x**2+2


print("x", x)


print("y", y)


print("z", z)
```

```python
plt.plot(x, y, 'green')  # plot a line using one line
```

```python
plt.subplot(1, 2, 1)  # plot subgraph using subplots(rows, columns, indexes)
plt.plot(x, y, 'r--')  # color and line style
plt.title("a")
plt.subplot(1, 2, 2)
plt.plot(x, z, 'g*--')
plt.title("b")
```

```python
plt.subplot(2, 1, 1)  # plot subgraph using subplots(rows, columns, indexes)
# third value takes color and line style , s Represents square markers
plt.plot(x, y, 'rs')
plt.subplot(2, 1, 2)
plt.plot(x, z, 'g--v')  # v Represents triangle-down markers
```

```python
fig, axes = plt.subplots(dpi=50)
axes.set_xlabel("x-axis")
axes.set_ylabel("y-axis")
axes.set_title("title of the graph")
axes.plot(x**2, 'r')
axes.plot(x*3, 'g')
# Makes legend of label-1 and label-2 in the graph
axes.legend(["label-1", "label-2"], loc=0)
```

```python
fig, axes = plt.subplots(dpi=100)
# colors in matplotlib
axes.plot(x, x+1, color='red')
axes.plot(x, x+2, color='green')
axes.plot(x, x+3, color='blue')
```

```python
fig, axes = plt.subplots(dpi=100)
# transparency of colors in matplotlib
axes.plot(x, x+1, color='red', alpha=1)
axes.plot(x, x+2, color='green', alpha=0.5)
axes.plot(x, x+3, color='blue', alpha=0.3)
```

```python
fig, axes = plt.subplots(dpi=100)
# linwidth and linestyles
axes.plot(x, x+1, color='red', linewidth=1, linestyle='--')
axes.plot(x, x+2, color='green', linewidth=10, linestyle='-.')
axes.plot(x, x+3, color='blue', linewidth=5, linestyle=':')
```

```python
fig, axes = plt.subplots(dpi=100)
# markers on line
axes.plot(x, x+1, color='red', marker='o')
axes.plot(x, x+2, color='green', marker='+')
axes.plot(x, x+3, color='blue', marker='s')
```

```python
fig, axes = plt.subplots(dpi=100)
# size of markers
axes.plot(x, x+1, color='green', marker='+', markersize=10)
axes.plot(x, x+2, color='green', marker='s', markersize=10)
axes.plot(x, x+3, color='green', marker='o', markersize=10)
```

```python
fig, axes = plt.subplots(dpi=100)
# markerfacecolor
axes.plot(x, x+1, color='green', marker='+',
          markersize=10, markerfacecolor='red')
axes.plot(x, x+2, color='green', marker='s',
          markersize=10, markerfacecolor='blue')
axes.plot(x, x+3, color='green', marker='o',
          markersize=10, markerfacecolor='red')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].plot(x, x**2, x, x**3, linewidth=2)
axes[1].plot(x, x**2, x, x**3, linewidth=2)
axes[1].set_ylim([1, 60])
axes[1].set_xlim([1, 5])
```

2D plots


```python
n = np.array([0, 1, 2, 3, 4, 5])
print(len(n))
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
axes[0].set_title("scatter plot")
axes[0].scatter(n, n+0.25*np.random.randn(len(n)), color='blue')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].set_title("scatter plot")
axes[0].scatter(n, n+0.25*np.random.randn(len(n)), color='blue')

axes[1].set_title("step plot")
axes[1].step(n, n**2, linewidth=2, color='blue')
```

```python
n = np.array([1, 2, 3, 4, 5, 6])
fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].set_title("scatter plot")
axes[0].scatter(n, n+0.25*np.random.randn(len(n)), color='blue')

axes[1].set_title("step plot")
axes[1].step(n, n**2, linewidth=2, color='blue')

axes[2].set_title("bar graph")
axes[2].bar(n, n**2, align="center", width=0.5, color='blue')

axes[3].set_title("fill area between curves")
axes[3].fill_between(n, n**2, n**3, color='blue')
```
